# 03 - Feature Engineering

```python
import pandas as pd
import numpy as np

df = pd.read_csv('telco_churn_cleaned.csv')
print(df.shape)
```

```python
# Drop identifier column (not predictive, keep a copy for reference if needed)
df = df.drop('customerID', axis=1, errors='ignore')
```

```python
# Encode target variable
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
# Feature creation: average monthly spend relative to tenure
df['AvgMonthlySpend'] = df['TotalCharges'] / df['tenure'].replace(0, 1)

# Feature creation: tenure buckets
df['TenureGroup'] = pd.cut(
    df['tenure'],
    bins=[0, 12, 24, 48, 60, np.inf],
    labels=['0-1yr', '1-2yr', '2-4yr', '4-5yr', '5yr+']
)
```

```python
# One-hot encode all remaining categorical columns (includes TenureGroup)
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Convert any bool columns to int
bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)

print(df.shape)
print(df.isnull().sum().sum(), 'missing values remaining')
```

```python
# Save feature-engineered dataset
df.to_csv('telco_churn_features.csv', index=False)
print('Saved: telco_churn_features.csv')
```
